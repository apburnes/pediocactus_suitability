````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
from osgeo import gdal, ogr
import time, os, StringIO

csv_file = "./pedio_cactus.csv"


col_names = [ 'fid', 'easting', 'northing', 'species', 'musym',
             'areasymbol', 'lru', 'esd', 'aspct', 'slope', 'dem', 
             'fall09', 'fall11', 'sprng09', 'sprng10', 'precip']

data_types = [np.int64, np.int64, object, object, object, object, object, float64, float64, 
              np.float64, np.float64, np.float64, np.float64, np.float64, np.float64]

data_dict = dict(zip(col_names, data_types))

data = pd.read_csv(csv_file, skiprows=1, names=col_names)
del data['fid']
subdata = pd.DataFrame(data=([data.aspct, data.slope, data.dem, data.precip, data.fall09, data.fall11, data.sprng09, data.sprng10]), 
                       dtype=float64).transpose()
````

````python
import itertools

flt_data = []

for col in subdata_cols:
    dx = subdata[col]
    if dx.dtype == float64:
        flt_data.append(col)

# Set empty datafram to the series length of the selected dataframe columns
framelen = subdata.shape[0]
new_data = pd.DataFrame(index=range(framelen))
new_data[flt_data] = subdata[flt_data]

d_x = subdata[flt_data]
dx_arr = [array(x) for x in d_x.values]
dxt = transpose(dx_arr)


def scatterplot_matrix(data, names, **kwargs):
    
    numlen, numvars = data[names].shape
    
    data_select = data[names]
    dx = transpose([array(x) for x in data_select.values])
    
    
    fig, axes = plt.subplots(nrows=numvars, ncols=numvars, figsize=(9,9))
    fig.subplots_adjust(hspace=0.05, wspace=0.05)
    
    for ax in axes.flat:
        
        ax.xaxis.set_visible(False)
        ax.yaxis.set_visible(False)
        
    for i, j in zip(*np.triu_indices_from(axes, k=1)):
        for x, y in [(i,j),(j, i)]:
            axes[x,y].plot(dx[x], dx[y], **kwargs)
    
    return fig


fig = scatterplot_matrix(subdata, flt_data, linestyle='none', marker='o', color='black', mfc='none')
plt.show()
````
