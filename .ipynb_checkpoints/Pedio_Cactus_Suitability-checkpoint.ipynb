````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
from osgeo import gdal, ogr
import time, os, json

s = json.load( open("./styles/bmh_matplotlibrc.json"))
matplotlib.rcParams.update(s)


csv_file = "./pedio_cactus.csv"


col_names = [ 'fid', 'easting', 'northing', 'species', 'musym',
             'areasymbol', 'lru', 'esd', 'aspct', 'slope', 'dem', 
             'fall09', 'fall11', 'sprng09', 'sprng10', 'precip']

data_types = [np.int64, np.int64, object, object, object, object, object, float64, float64, 
              np.float64, np.float64, np.float64, np.float64, np.float64, np.float64]

data_dict = dict(zip(col_names, data_types))

data = pd.read_csv(csv_file, skiprows=1, names=col_names)
del data['fid']

````

````python
def col_avg(x):
    return pd.Series((x['fall09']+x['fall11'] + x['sprng09'] + x['sprng10'])/4)

data['cvr_avg'] = data.apply(col_avg, axis=1)

sub_cols = ['aspct', 'slope', 'dem', 'precip', 'cvr_avg']

subdata = data[sub_cols]

##convert cover avrerage to float
subdata['cvr_avg'] = subdata['cvr_avg'].astype(float64)

from scipy.cluster.vq import *
from sklearn import cluster, datasets

k_means = cluster.KMeans(k=3)
k_means.fit(subdata)

k_clust = k_means.labels_[0:]
subdata['cluster'] = pd.Series(k_clust)



print range(max(subdata.index)+1)

````

````python
#### Use column names to iterate through matrix
sub_z_col = []

for i, col in enumerate(sub_cols):
    if col != 'cluster':
        z_col = col + "_z"
        
        col_mean = subdata[col].mean()
        col_std = subdata[col].std()
    
        print col
        print "Mean = %s" % (col_mean)
        print "STD = %s" % (col_std)
        
        
        sub_z_col.append(z_col)
        sub_d = subdata
        
        sub_d[z_col] = stats.zscore(subdata[col])
        max_z = sub_d[z_col].max()
        min_z = sub_d[z_col].min()
        max_ix = sub_d[z_col].argmax()
        min_ix = sub_d[z_col].argmin()
        max_value = sub_d[col][max_ix]
        min_value = sub_d[col][min_ix]
        print z_col
        print "Max| index: %s @ value: %s | std: %s" % (max_ix, max_value, max_z)
        print "Min| index: %s @ value: %s | std: %s \n" % (min_ix, min_value, min_z)


sub_z = sub_d[sub_z_col]
sub_z.head()
````
